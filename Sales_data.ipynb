```python
import pandas as pd
import matplotlib.pyplot as plt


sales_file = pd.read_csv("100 Sales Records.csv.csv")
sales_file = sales_file.head(30)
grp = sales_file.groupby('Item Type')

summary = pd.DataFrame({
    'Units Sold': grp['Units Sold'].sum(),
    'Total Revenue': grp['Total Revenue'].sum(),
    'Total Profit': grp['Total Profit'].sum()
})

summary.plot(kind='line', marker='o', figsize=(12,6))

plt.title("Sales Summary")
plt.xlabel("Item Type")
plt.ylabel("Values")
plt.xticks(rotation=45)
plt.legend(title="Sales data")
plt.tight_layout()
plt.show()
```
